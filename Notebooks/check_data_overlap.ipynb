```python
import pandas as pd
import numpy as np
from rdkit import Chem
import h5py
```

## Assay data

```python
df = pd.read_csv('/projects/global/gred/resbioai/multimodal/PUMA_data/assay_data/assay_matrix_discrete_270_assays.csv')
```

```python

def get_canonical_SMILES(s):
    return Chem.MolToSmiles(Chem.MolFromSmiles(s))

df['valid_smiles'] = [get_canonical_SMILES(s) for s in df['smiles']]
```

```python
mols1 = df['valid_smiles'].values
mols1
```
